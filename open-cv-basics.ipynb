```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('Computer-Vision-with-Python/DATA/00-puppy.jpg')
type(img)
```

```python
img.shape
```

```python
plt.imshow(img) #why blue?, opencv and matplot have different color channel orders
#plt -->  RGB
#opencv --> BGR
```

```python
#transform image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#display 'fixed' image
plt.imshow(img)
```

```python
fix_img = img.copy()
fix_img.shape
```

```python
new_img = cv2.resize(fix_img, (1000, 400)) #opposite of numpy array
plt.imshow(new_img)
```

```python
w_ratio = .75
h_ratio = .75
```

```python
new_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio) #resize based on ratio
plt.imshow(new_img)
```

```python
#flip on hor or ver axis
new_img = cv2.flip(fix_img, 0) #hor flip
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, 1) #ver flip
plt.imshow(new_img)
```

```python
cv2.imwrite('totally_new.jpg', fix_img) #writes to disk
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
blank_img.shape
```

```python
cv2.rectangle(blank_img, pt1=(200, 100), pt2=(300, 150), color=(0, 255, 0), thickness=5)
#pt1=top l pt2= bottom r
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(256, 256), pt2=(356, 356), color=(0, 0, 255), thickness=7)
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img, center=(100, 400), radius=50, color=(255, 0, 0), thickness=-1)
plt.imshow(blank_img)
#thickness -1 is a filled shape
```

```python
cv2.line(blank_img, pt1=(150,100), pt2=(300,300), color=(102, 255, 255), thickness=8)
plt.imshow(blank_img)
#pt1 x axis y axis top left 
#pt 2 x axis y axis bottom right
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text='Hello Family!', org=(10, 450), fontFace=font, fontScale=2, color=(255,255,255),
            thickness=3, lineType=cv2.LINE_AA)

plt.imshow(blank_img)
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(blank_img)
```

```python
vertices = np.array([ [100,300],[200, 200],[400, 300],[200,400] ], dtype=np.int32)
```

```python
vertices.shape #opencv wants 3 dimensions   
```

```python
pts = vertices.reshape((-1, 1, 2))
```

```python
pts.shape #3rd dimension added
```

```python
pts
```

```python
cv2.polylines(blank_img, [pts], isClosed=True, color=(255,0,0), thickness=5) #need to pass reshape as list
plt.imshow(blank_img)
```
